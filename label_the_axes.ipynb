```python
# import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load data from csv file into dataframe

campaigns = pd.read_csv("campaigns_cleaned.csv", parse_dates=["Reporting starts", "Reporting ends"])
```

```python
# display the first few rows of the data

campaigns.head()
```

```python
# Calculate revenue from mobile app installs and add as a new column

campaigns["Revenue from mobile app installs"] = campaigns["Mobile app installs"] * campaigns["Cost per mobile app install"]

# group by Campaign and get mean revenue from mobile app installs for 10 campaigns

campaigns_grouped = campaigns.groupby(by="Campaign")["Revenue from mobile app installs"].mean().iloc[15:25]

# create a bar plot

campaigns_grouped.plot.bar()

# label the x-axis

plt.xlabel("Campaign")

# label the y-axis

plt.ylabel("Average revenue from mobile app installs")

# display the plot

plt.show()
```

```python
# Calculate revenue from mobile app installs and add as a new column

campaigns["Revenue from mobile app installs"] = campaigns["Mobile app installs"] * campaigns["Cost per mobile app install"]

# group by Campaign and get mean revenue from mobile app installs for 10 campaigns

campaigns_grouped = campaigns.groupby(by="Campaign")["Revenue from mobile app installs"].mean().iloc[15:25]

# create a bar plot

campaigns_grouped.plot.bar()

# create the x tick positions

x_tick_positions = np.arange(0,10)

# get the campaign numbers

campaign_numbers = campaigns_grouped.index.values

# create the x tick labels

x_tick_labels = ["Campaign"+str(label) for label in campaign_numbers]

# label the x ticks and rotate them by 45 degrees horizontally to the right

plt.xticks(ticks=x_tick_positions, labels=x_tick_labels, rotation=45, ha="right")

# label the x-axis

plt.xlabel("Campaign")

# label the y-axis

plt.ylabel("Average revenue from mobile app installs")

# display the plot

plt.show()
```
