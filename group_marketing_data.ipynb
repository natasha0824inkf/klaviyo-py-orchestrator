```python
# import relevant libraries

import pandas as pd
```

```python
# load data from csv file into dataframe

data = pd.read_csv("channels.csv")
```

```python
# view first few rows

data.head()
```

```python
# get some basic information

data.info()
```

```python
# group data by "Channel" and get average "Sales" per group 

data[["Channel", "Sales"]].groupby("Channel").mean()
```

```python
# group data by "Region" and get average "Sales" per group 

data[["Region", "Sales"]].groupby("Region").mean()
```

```python
# group data by "Channel" and "Region" and get average "Sales" and average "Clicks" per group 

data.groupby(["Channel", "Region"]).mean()
```

```python
# group data by "Channel" and "Region" and get total "Sales" and total "Clicks" per group 

data.groupby(["Channel", "Region"]).sum()
```
