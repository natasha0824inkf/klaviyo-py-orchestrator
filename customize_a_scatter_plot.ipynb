```python
# import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data from csv file into dataframe

data = pd.read_csv("analytics_search_data.csv")
```

```python
# display the first few rows of the data

data.head()
```

```python
# specify figure size

plt.figure(figsize=(15,5))

# create scatter plot

sns.scatterplot(data=data, x="impressions", y="clicks")

# display plot

plt.show()
```

```python
# specify figure size

plt.figure(figsize=(15,5))

# create scatter plot, customize the hue and size of the data points

sns.scatterplot(data=data, x="impressions", y="clicks", hue="position", size="views")

plt.show()
```
