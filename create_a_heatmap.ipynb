```python
# import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data from csv file into dataframe

data = pd.read_csv("social_media_campaign_data.csv")
```

```python
# display the first few rows of the data

data.head()
```

```python
# use pandas to create a correlation matrix and store it in a variable

data_corr = data.corr()
data_corr
```

```python
# use matplotlib to specify the size of the figure

plt.figure(figsize=(20,16))

# use seaborn to create the heatmap

sns.heatmap(data_corr, #load the correlation matrix
           linewidth=0.5, #line width
           cmap='RdBu_r', #mapping from data values to color space 
           vmin=-1, #minimum value to anchor the colormap
           vmax=1, #maximum value to anchor the colormap
           annot=True) #annotate the cells of the heatmap

# show the plot

plt.show() 
```
