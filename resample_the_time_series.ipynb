```python
# import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data from csv file into dataframe

page_traffic = pd.read_csv("page_traffic_updated.csv", parse_dates=["date"], index_col="date")

# view first few rows

page_traffic.head()
```

```python
# resample monthly

page_traffic_monthly = page_traffic.resample("M").mean()

# check it out 

page_traffic_monthly
```

```python
# visualize resampled monthly page traffic data 

# specify figure size

plt.figure(figsize=(15,5))

# create line plot

sns.lineplot(data=page_traffic_monthly, x="date", y="views", marker="o")

# add title

plt.title("Line plot: Resampled monthly webpage traffic")

# add grid

plt.grid(True)

# display plot

plt.show()
```

```python
# resample quarterly

page_traffic_quarterly = page_traffic.resample("Q").mean()

# check it out 

page_traffic_quarterly
```

```python
# visualize resampled quarterly page traffic data 

# specify figure size

plt.figure(figsize=(15,5))

# create line plot

sns.lineplot(data=page_traffic_quarterly, x="date", y="views", marker="o")

# add title

plt.title("Line plot: Resampled quarterly webpage traffic")

# add grid

plt.grid(True)

# display plot

plt.show()
```
