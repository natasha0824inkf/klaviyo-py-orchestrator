```python
# import relevant libraries

import pandas as pd
```

```python
# load data from csv file into dataframe

data = pd.read_csv("analytics_search_data.csv")
```

```python
# view first few rows of data

data.head()
```

```python
# filter and sort the data

# filter the data by pages that have received 100 or more impressions, then sort by exit rate in ascending order, then save resulting dataframe in a variable

filtered_data = data[data["impressions"] >= 100].sort_values(by="exit_rate", ascending=True)
```

```python
# view first few rows of filtered_data

filtered_data.head()
```

```python
# export the filtered_data to a CSV file which will populate into your local file system

filtered_data.to_csv("filtered_data.csv", index=False)
```
