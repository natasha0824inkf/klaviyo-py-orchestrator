```python
# import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data from csv file into dataframe

data = pd.read_csv("analytics_search_data.csv")
```

```python
# display the first few rows of the data

data.head()
```

```python
# set the dataframe index to the "page" column and then remove the "page" column
# save the result in a new variable

data_updated = data.set_index("page", drop=True)
```

```python
# check it out

data_updated.head()
```

```python
# create subplots

fig, axs = plt.subplots(1, 2, figsize=(15,5))

# create bar plot of clicks for first 10 pages

data_updated["clicks"][0:10].plot.bar(ax=axs[0])
axs[0].set_xlabel("Page")
axs[0].set_ylabel("Clicks")
axs[0].set_title("Bar plot of clicks for first 10 pages")

# create bar plot of views for first 10 pages

data_updated["views"][0:10].plot.bar(ax=axs[1])
axs[1].set_xlabel("Page")
axs[1].set_ylabel("Views")
axs[1].set_title("Bar plot of views for first 10 pages")

# show the plots

plt.show()
```
