```python
# import relevant libraries

import pandas as pd
```

```python
# load data from csv file into dataframe

analytics_data = pd.read_csv("analytics_data.csv")
```

```python
# view the data that's been loaded in

analytics_data.head()
```

```python
# get the current column names in analytics_data

analytics_data.columns
```

```python
# rename the columns in analytics_data as data cleaning step

analytics_data = analytics_data.rename(columns = {"URL":"url", "Pageviews":"views", "Unique Pageviews":"unique_views", "Avg. Time on Page":"avg_time", 
                                                  "Entrances":"entrances", "Bounce Rate":"bounce_rate", "% Exit":"exit_rate", "Page Value":"value"})

# display the first few rows of analytics_data after the update

analytics_data.head()
```

```python
# insert the missing part of the URL to every value in the "url" column 

analytics_data["url"] = "https://" + analytics_data["url"]

# display the first few rows of analytics_data after the update

analytics_data.head()
```
