```python
# import relevant libraries

import warnings
import pandas as pd

# suppress FutureWarning messages
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# load data from csv file into dataframe

analytics_data = pd.read_csv("analytics_data_cleaned.csv")
```

```python
# view the data that's been loaded in

analytics_data.head()
```

```python
# check the data types in analytics_data

analytics_data.dtypes
```

```python
# convert avg_time column from string to time delta

analytics_data["avg_time"] = pd.to_timedelta(analytics_data["avg_time"])

# display first few rows of analytics_data after the update

analytics_data.head()
```

```python
# replace % with an empty string in bounce_rate

analytics_data["bounce_rate"] = analytics_data["bounce_rate"].str.replace("%","")

# convert bounce_rate column to float

analytics_data["bounce_rate"] = analytics_data["bounce_rate"].astype("float")

# display first few rows of analytics_data after the update

analytics_data.head()
```

```python
# replace % with an empty string in exit_rate

analytics_data["exit_rate"] = analytics_data["exit_rate"].str.replace("%","")

# convert exit_rate column to float

analytics_data["exit_rate"] = analytics_data["exit_rate"].astype("float")

# display first few rows of analytics_data after the update

analytics_data.head()
```

```python
# replace % with an empty string in value column

analytics_data["value"] = analytics_data["value"].str.replace("$","")

# convert value column to float

analytics_data["value"] = analytics_data["value"].astype("float")

# display first few rows of analytics_data after the update

analytics_data.head()
```
