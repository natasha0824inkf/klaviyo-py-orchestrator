```python
# load datasets from csv files into dataframes

pages_visited = pd.read_csv("pages_visited.csv")

campaigns_invested = pd.read_csv("campaigns_invested.csv")
```

```python
# calculate conversion rate, as a percentage, for the pages_visited data
# add the result as a new column 

pages_visited["conversion_rate"] = (pages_visited["conversions"] / pages_visited["visits"]) * 100
```

```python
# calculate return on investment, or ROI, as a percentage, for the campaigns_invested data
# add the result as a new column 

campaigns_invested["ROI"] = ((campaigns_invested["revenue"] - campaigns_invested["investment"]) / campaigns_invested["investment"]) * 100
```
