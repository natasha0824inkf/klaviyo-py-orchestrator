```python
# import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data from csv file into dataframe

page_traffic = pd.read_csv("page_traffic_updated.csv", parse_dates=["date"], index_col="date")

# view first few rows

page_traffic.head()
```

```python
# create a rolling 7-day window, calculate the means, and save in a variable

rolling_7d = page_traffic["views"].rolling(window=7, # set a rolling window of 7 days
                                           center=True # label each window at its midpoint
                                           ).mean()
rolling_7d
```

```python
# visualize 7-day rolling averages

# specify figure size

plt.figure(figsize=(15,5))

# create a plot of the rolling averages

sns.lineplot(rolling_7d)

# add title

plt.title("7-day rolling average plot: webpage traffic")

# display the plot

plt.show()
```

```python
# visualize 30-day rolling averages

# specify figure size

plt.figure(figsize=(15,5))

# create a plot of the rolling averages

sns.lineplot(rolling_30d)

# add title

plt.title("30-day rolling average plot: webpage traffic")

# display the plot

plt.show()
```
