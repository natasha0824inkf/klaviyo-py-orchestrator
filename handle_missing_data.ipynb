```python
# import relevant libraries

import pandas as pd
```

```python
# load data from csv file into dataframe

campaigns = pd.read_csv("campaigns.csv", parse_dates=["Reporting starts", "Reporting ends"])
```

```python
# display the first few rows of campaigns

campaigns.head()
```

```python
# count the number of missing values per column

campaigns.isna().sum()
```

```python
# count how many rows have missing values

len(campaigns) - len(campaigns.dropna())
```

```python
# total rows

len(campaigns)
```

```python
# fill missing values with zero, in place

campaigns.fillna(0, inplace=True)

# display first few rows of campaigns after the update

campaigns.head()
```
