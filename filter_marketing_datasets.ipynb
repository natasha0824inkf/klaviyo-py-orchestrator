```python
# import relevant libraries

import pandas as pd
```

```python
# load data from csv file into dataframe

data = pd.read_csv("analytics_search_data.csv")
```

```python
# view first few rows

data.head()
```

```python
# filter the data by pages that have received 100 or more clicks

data[data["clicks"] >= 100]
```

```python
# filter the data by pages that have received 100 or more clicks, extracting a subset of columns

data.loc[data["clicks"] >= 100, ["page", "clicks", "impressions", "unique_views"]]
```

```python
# filter with multiple conditions

# filter the data by pages that have received 100 or more impressions *and* position 10 or less

data[(data["impressions"] >= 100) & (data["position"] <= 10)]
```

```python
# filter and sort the data

# filter the data by pages that have received 100 or more impressions, then sort by exit rate in ascending order

data[data["impressions"] >= 100].sort_values(by="exit_rate", ascending=True)
```
