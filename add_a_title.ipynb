```python
# import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data from csv file into dataframe

campaigns = pd.read_csv("campaigns_cleaned.csv", parse_dates=["Reporting starts", "Reporting ends"])
```

```python
# display the first few rows of the data

campaigns.head()
```

```python
# Calculate revenue from mobile app installs and add as a new column

campaigns["Revenue from mobile app installs"] = campaigns["Mobile app installs"] * campaigns["Cost per mobile app install"]

# group by Campaign and get mean revenue from mobile app installs for 10 campaigns

campaigns_grouped = campaigns.groupby(by="Campaign")["Revenue from mobile app installs"].mean().iloc[15:25]

# create a bar plot

campaigns_grouped.plot.bar()

# label the x-axis

plt.xlabel("Campaign")

# label the y-axis

plt.ylabel("Average revenue from mobile app installs")

# add a title to the plot

plt.title("Average revenue from mobile app installs across 10 campaigns")

# display the plot

plt.show()
```
