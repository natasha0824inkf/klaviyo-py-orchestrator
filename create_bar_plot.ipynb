```python
# import relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data from csv file into dataframe

campaigns = pd.read_csv("campaigns_cleaned.csv", parse_dates=["Reporting starts", "Reporting ends"])
```

```python
# display the first few rows of the data

campaigns.head()
```

```python
# Calculate revenue from mobile app installs and add as a new column

campaigns["Revenue from mobile app installs"] = campaigns["Mobile app installs"] * campaigns["Cost per mobile app install"]

# group by Campaign and create a bar plot to visualize mean revenue from mobile app installs for 10 campaigns

campaigns.groupby(by="Campaign")["Revenue from mobile app installs"].mean().iloc[15:25].plot.bar()

# display the plot

plt.show()
```
